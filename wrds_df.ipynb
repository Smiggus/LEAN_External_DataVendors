5. Query Data from WRDS Server
- get_table() method
- straightforward if getting data from one single dataset
- specify which library/database and table/dataset to "get"
- can slice data by:
        number of rows
        column names

```python
# Sample to query data from WRDS server
```

```python
# Extract first 5 obs from comp.company

company = conn.get_table(library='comp', table='company', obs=5)
company.shape
```

```python
company
```

```python
# Uses permno sample
apple = conn.raw_sql("""select permno, date, prc, ret, shrout 
                        from crsp.msf 
                        where permno = 14593
                        and date>='01/01/2019'""", 
                     date_cols=['date'])
apple
```

```python
# Table Values:
'''
    tic : ticker
    busdesc : business description
    cusip : cusip
    loc : current iso country code (HQ)
    gind : gics industries
    gsector : gics sector
    div : dividend or divd : cash dividends daily
    prccd : price close daily
    prchd : price high daily
    prcld : price low daily
    prcod : price open daily
    trfd : daily total return factor
'''
apple = conn.raw_sql("""select tic
                        from comp_na_daily_all.security 
                        where tic = 'AAPL'
                        """, 
                     date_cols=['date'])
```
